```python
import csv
from datetime import datetime
from os import path
import pandas as pd 
```

```python
def Ensure_file():

    file_name = "Expenses.csv"

    if not path.exists(file_name):
        with open(file_name,mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["Date", "Category", "Description", "Amount"])
```

```python
def get_amount():
    
    user = input("Amount (in â‚¹): ").strip()
    try:
        amt = int(user)
        return amt
    except Exception:
        print("Enter a valid number!") 
        return get_amount()
```

```python

def Add_Expenses():
    date = datetime.now().strftime("%d/%m/%Y")
    Categories = input(f"Choose the type of expense ( Food / Bills / Travel / Fun / Others))").title().strip()
    if Categories.lower() in ["food", "travel", "bills", "fun", "others"]:
        category = Categories
    else: 
        print("Please select one from ( Food / Bills / Travel / Fun / Others) ")

    def get_description():
        note = input("Description: ").title()
        if len(note) <=25:
            return note
             
        else: 
            print("Too long. Keep it under 25 characters.")
            get_description()
            
    description = get_description()

    amount = get_amount()

    with open("Expenses.csv", mode="a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([date, category, description, amount])
        
    print("Expenses added successfully!")

```
